```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rk(x0, y0, f, l, n):
    h = l / n
    grid = dict()
    grid[x0] = y0
    for i in range(0, n):
        xi = x0 + i * h
        yi = grid[xi]
        fi = f(xi, yi)
        xs = x0 + (i + 1) * h
        ys = yi + h * (fi + f(xi + h, yi + h * fi)) / 2
        grid[xs] = ys
    return grid
        
    
    
```

```python
def f(x, y):
    if y == 1:
        return (x - x ** 2) * y
        return x * y * (1 - x)
    else:
        return 1
    
f = lambda x, y: (x * y * (1 - x)) if (y != 0) else 1
    
    

n = 40
x0 = 0
y0 = 1
l = 5
g = rk(x0, y0, f, l, n)


x = list(g)
y = [g[xx] for xx in x]
plt.scatter(x, y)
```
